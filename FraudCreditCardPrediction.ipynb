```python
# Importing the important dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# To load the dataset to a pandas dataframe
credit_card_data=pd.read_csv('/content/sample_data/creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# Dataset Information
credit_card_data.info()
```

```python
# Checking the number of missing value in each column
credit_card_data.isnull().sum
```

```python
# Distribution of legit transactions and fraudulent transactions
# 0 --> Normal Transactions
# 1 --> Fraudulent Transactions
# Highly Unbalanced Dataset
credit_card_data['Class'].value_counts()
```

```python
# Separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical Measure of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the value between both classes by their mean value
credit_card_data.groupby('Class').mean()
```

```python
# Build a sample dataset containing similar distribution of original dataset for making uniform distribution
legit_sample = legit.sample(n=492) # Randomly sampling the dataset
```

```python
# Concatenating two Dataframes
new_dataset = pd.concat([legit_sample, fraud],axis=0) # Axis 0 means row
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# Split the dataset into features and target(0 or 1)
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
print(Y)
```

```python
# Split the dataset into training and testing data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# To train our model
model = LogisticRegression()
```

```python
# Training the logistic regression model by using training data
model.fit(X_train, Y_train)
```

```python
# To evaluate our model based on accuracy score
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training data: ", training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on testing data: ", testing_data_accuracy)
```
